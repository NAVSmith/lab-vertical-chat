# Vertical Chat
A sample how to build a chat for small business using:

* GPT 4 mini
* Panel
* OpenAI


This is just a simple sample to start to understand how the OpenAI API works, and how to create Prompts. It Is really far from beign a complete solution.
We are going to introduce some interesting points:

* The roles in a conversation.
* How is the conversations’ memory preserved?

Deeper explanations in the article: [Create Your First Chatbot Using GPT 3.5, OpenAI, Python and Panel.](https://medium.com/towards-artificial-intelligence/create-your-first-chatbot-using-gpt-3-5-openai-python-and-panel-7ec180b9d7f2)

```python
#if you need an API Key from OpenAI
#https://platform.openai.com/account/api-keys

from openai import OpenAI
import os

from dotenv import load_dotenv, find_dotenv
_ = load_dotenv(find_dotenv('../../open_ai.env')) # read the env file from parent directory

OPENAI_API_KEY  = os.getenv('OPENAI_API_KEY')
```

```python
client = OpenAI(
    # This is the default and can be omitted
    api_key=OPENAI_API_KEY,
)

def continue_conversation(messages, temperature=0):
    response = client.chat.completions.create(
        model="gpt-4o-mini",
        messages=messages,
        temperature=temperature,
    )
    #print(str(response.choices[0].message["content"]))
    return response.choices[0].message.content
```

```python
#Creating the prompt
#read and understand it.
import panel as pn  # GUI

#Creating the panel.
pn.extension()


panels = []

client_prompt = pn.widgets.TextInput(value="Hi", placeholder='Enter text here…')
button_conversation = pn.widgets.Button(name="talk")


context = [ {'role':'system', 'content':"""
Act as an OrderBot, you work collecting orders in a delivery only fast food restaurant called
My Dear Frankfurt. \
First welcome the customer, in a very friendly way, then collects the order. \
You wait to collect the entire order, beverages included \
then summarize it and check for a final \
time if everything is ok or the customer wants to add anything else. \
Finally you collect the payment.\
Make sure to clarify all options, extras and sizes to uniquely \
identify the item from the menu.\
You respond in a short, very friendly style. \
The menu includes \
burger  12.95, 10.00, 7.00 \
frankfurt   10.95, 9.25, 6.50 \
sandwich   11.95, 9.75, 6.75 \
fries 4.50, 3.50 \
salad 7.25 \
Toppings: \
extra cheese 2.00, \
mushrooms 1.50 \
martra sausage 3.00 \
canadian bacon 3.50 \
romesco sauce 1.50 \
peppers 1.00 \
Drinks: \
coke 3.00, 2.00, 1.00 \
sprite 3.00, 2.00, 1.00 \
vichy catalan 5.00 \
"""} ]



def add_prompts_conversation(_):
    #Get the value introduced by the user
    prompt = client_prompt.value_input
    client_prompt.value = ''

    #Append to the context the User prompt.
    context.append({'role':'user', 'content':f"{prompt}"})

    #Get the response.
    response = continue_conversation(context)

    #Add the response to the context.
    context.append({'role':'assistant', 'content':f"{response}"})

    #Update the panels to show the conversation.
    panels.append(
        pn.Row('User:', pn.pane.Markdown(prompt, width=600)))
    panels.append(
        pn.Row('Assistant:', pn.pane.Markdown(response, width=600)))

    return pn.Column(*panels)


interactive_conversation = pn.bind(add_prompts_conversation, button_conversation)
```

```python

dashboard = pn.Column(
    client_prompt,
    pn.Row(button_conversation),
    pn.panel(interactive_conversation, loading_indicator=True),
)

dashboard
```

```python
import panel as pn  # GUI

#Creating the panel.
pn.extension()


panels = []

client_prompt = pn.widgets.TextInput(value="Hi", placeholder='Enter text here…')
button_conversation = pn.widgets.Button(name="talk")


context = [
  {
    'role': 'system',
    'content': """
Act as a ReceptionDeskBot for a large enterprise company. \
You are the first point of contact for visitors, callers, vendors, and employees. \
Begin every interaction with a professional, warm, and welcoming greeting. \
Your primary goals are security, efficiency, and excellent visitor experience. \
Politely identify the visitor or caller’s name, company (if applicable), and purpose of visit. \
Determine the visitor type: employee, scheduled guest, walk-in visitor, vendor, contractor, delivery, or interview candidate. \
Always follow company security protocols and never grant access without proper authorization. \
If required, request and verify identification and confirm appointments before proceeding. \
Notify the host or department using approved internal communication channels. \
If the host is unavailable, offer appropriate alternatives such as waiting, contacting a delegate, or rescheduling. \
Provide clear directions or instructions and explain next steps in simple language. \
For phone calls, quickly identify the caller’s intent and route the call to the correct department or individual. \
Never disclose confidential information, employee availability, or internal details without authorization. \
Remain calm, respectful, and professional at all times, including in stressful or unexpected situations. \
If an issue cannot be resolved at the reception level, escalate according to protocol. \
Before ending any interaction, summarize what was done and ask if further assistance is needed. \
Respond in a concise, polite, and professional style suitable for a corporate front desk environment.
"""
  }
]


def add_prompts_conversation(_):
    #Get the value introduced by the user
    prompt = client_prompt.value_input
    client_prompt.value = ''

    #Append to the context the User prompt.
    context.append({'role':'user', 'content':f"{prompt}"})

    #Get the response.
    response = continue_conversation(context)

    #Add the response to the context.
    context.append({'role':'assistant', 'content':f"{response}"})

    #Update the panels to show the conversation.
    panels.append(
        pn.Row('User:', pn.pane.Markdown(prompt, width=600)))
    panels.append(
        pn.Row('Assistant:', pn.pane.Markdown(response, width=600)))

    return pn.Column(*panels)


interactive_conversation = pn.bind(add_prompts_conversation, button_conversation)
```

Some prompt variations that were too open-ended or overly conversational caused issues. In these cases, the chatbot occasionally:
- Assumed visitor authorization without explicit confirmation
- Hallucinated internal roles, departments, or employee availability
- Offered directions or access instructions without verifying security requirements
- Attempted to resolve situations that should have been escalated
